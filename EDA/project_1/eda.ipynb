```python
import pandas as pd 
import plotly.express as px
import plotly.graph_objects as go 
```

```python
df=pd.read_csv("./data/vgsales.csv")
df.head()
```

```python
#- Which year had the highest sales?
df_copy=df.groupby("Year")["Global_Sales"].sum().reset_index()
fig=px.line(
    df_copy,
    x="Year",
    y="Global_Sales",
    markers=True
)

fig.show()
```

```python
 #-  Do any consoles seem to specialize in a particular genre?
coppy_df=df.groupby(["Platform","Genre"])["Global_Sales"].sum().reset_index()
coppy_df=coppy_df.sort_values("Global_Sales",ascending=False)
fig=px.bar(
    coppy_df,
    x="Platform",
    y="Global_Sales",
    color="Genre",
    barmode="overlay"
)
fig.show()
```

```python
coppy_df=df
batida_hit_japonya_flop = coppy_df[(coppy_df['NA_Sales'] > 3.0) & (coppy_df['JP_Sales'] == 0)]
coppy_df=batida_hit_japonya_flop.sort_values('Global_Sales', ascending=False).head(5)

coppy_df['NA_JP_Fark'] = coppy_df['NA_Sales'] - coppy_df['JP_Sales']
en_buyuk_farklar = coppy_df.sort_values('NA_JP_Fark', ascending=False).head(10)

fig = px.bar(en_buyuk_farklar, 
             x='Name', 
             y=['NA_Sales', 'EU_Sales', 'JP_Sales'], # Üç bölgeyi de listeyle veriyoruz!
             title="Hit vs Flop",
             barmode='group') # Sütunları yan yana koy demek

fig.show()
```
