```python
#step3
"""

StudentID                    40drop x
Name                         34drop x
Gender                       48mod
AttendanceRate               40 x
StudyHoursPerWeek            50x
PreviousGrade                33
ExtracurricularActivities    43
ParentalSupport              22
FinalGrade                   40 medyan /mean x
Study Hours                  24medyan/ mean x
Attendance (%)               41
Online Classes Taken         25

"""
```

```python
#drop
st_per.dropna(subset=["StudentID","Name"],inplace=True)
#fill


st_per['AttendanceRate'] = st_per['AttendanceRate'].fillna(
    st_per.groupby('Online Classes Taken')['AttendanceRate'].transform('median')
)
st_per["FinalGrade"]=st_per["FinalGrade"].fillna(st_per["FinalGrade"].mean())
st_per["Study Hours"]=st_per["Study Hours"].fillna(st_per["Study Hours"].mean())
st_per["StudyHoursPerWeek"]=st_per["StudyHoursPerWeek"].fillna(st_per.groupby("FinalGrade")["StudyHoursPerWeek"].transform("median"))
st_per["AttendanceRate"]=st_per["AttendanceRate"].fillna(st_per["AttendanceRate"].median())
```

```python
st_per["PreviousGrade"]=st_per["PreviousGrade"].fillna(st_per.groupby("FinalGrade")["PreviousGrade"].transform("median"))
```

```python
st_per["ExtracurricularActivities"]=st_per["ExtracurricularActivities"].fillna(0.0)
```

```python
probs=st_per["ParentalSupport"].value_counts(normalize=True)
missing_mask = st_per['ParentalSupport'].isnull()
missing_count = missing_mask.sum()

random_values = np.random.choice(probs.index, size=missing_count, p=probs.values, replace=True)
st_per.loc[missing_mask, "ParentalSupport"] = random_values
```
