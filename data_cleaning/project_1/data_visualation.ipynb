```python
import pandas as pd
import plotly.express as px
```

```python
st_per=pd.read_csv("./clean_student_performance.csv")
st_per.head(10)
```

```python
fig = px.scatter(
    st_per, 
    x="StudyHoursPerWeek", 
    y="FinalGrade", 
    trendline="ols",             
    title="Çalışma Saati ve Final Notu İlişkisi",
    labels={"StudyHoursPerWeek": "Haftalık Çalışma Saati", "FinalGrade": "Final Notu"},
    color="Gender",           
    hover_data=['Name']          

fig.show()
```

```python
st_per['Online_Status'] = st_per['Online Classes Taken'].map({True: 'Alıyor', False: 'Almıyor'})
px.box(
    st_per,
    x="Online_Status",
    y="FinalGrade",
    title="Online Dersin Nota Etkisi",
    labels={"FinalGrade":"Final Grade"},
    color="Online Classes Taken",
)
```

Box plot analysis reveals no significant correlation between the number of online classes taken and final grades. This implies that online classes alone do not drive performance, suggesting that other factors—most notably StudyHoursPerWeek—have a greater impact on student success

```python
support_group = st_per.groupby('ParentalSupport')['StudyHoursPerWeek'].mean().reset_index()
fig=px.bar(
    support_group,
    x="ParentalSupport",
    y="StudyHoursPerWeek",
    color="ParentalSupport",
    title="Aile Desteğine Göre Ortalama Haftalık Çalışma Süresi"
)
fig.show()
```
