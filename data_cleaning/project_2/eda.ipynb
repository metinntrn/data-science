```python
#Soru 1: Katılımcıların en çok tercih ettiği sosyal medya platformları nelerdir? (Hangi platformun kullanıcı sayısı daha fazla?)
#Önerilen Grafik: Bar Plot (Çubuk Grafik) veya Pie Chart (Pasta Grafik)
most_used_platforms=df.platform_usage.value_counts()
most_used_platforms=pd.DataFrame(most_used_platforms).reset_index()
fig=px.bar(
    most_used_platforms,
    y="platform_usage",
    x="count",
    orientation="h",
    color="platform_usage",
    color_discrete_sequence=["#e6e600","#7d8887","#380e0e"],
    barmode="stack",
    text="count"
)
fig.update_traces(
    marker_line_color="black",
    marker_line_width=2,
    textposition="outside",

)
fig.update_yaxes(tickangle=45)
fig.update_layout(
    yaxis_title="Platform Names",
    xaxis_title="User Count",
    title="Most Used Plotforms", 
    height=400,


)
fig.show()
```

```python
#Soru 2: Veri setindeki katılımcıların yaş dağılımı nasıl bir eğilim gösteriyor? Gençler mi ağırlıkta yoksa daha homojen bir dağılım mı var?
#Önerilen Grafik: Histogram
age_distribution=df.age.value_counts()
age_distribution=pd.DataFrame(age_distribution).reset_index()
age_distribution['age_label'] = "Age " + age_distribution['age'].astype(str)
fig=px.pie(
    age_distribution,
    names="age_label",
    values="count",
    hole=0.4,
    color="count",
    title="Age Distribution",
)
fig.update_traces(
    textposition="inside",
    textinfo="label+percent",
    pull=[0,0,0,0.3,0,0],
    marker=dict(line=dict(color='black', width=1.5))
)
fig.update_layout(
    paper_bgcolor="#60818e",
    title_font_color="lightcyan",
    title_font_size=24,
    font_color="white"
)
fig.show()
```

```python
#Soru 3: Günlük sosyal medya kullanım süresi (daily_social_media_hours) arttıkça uyku süresi (sleep_hours) azalıyor mu?
#Önerilen Grafik: Scatter Plot (Dağılım Grafiği) - İstersen içine bir trend çizgisi (regresyon) de ekleyebilirsin.
fig=px.scatter(
    df,
    x="daily_social_media_hours",
    y="sleep_hours",
    marginal_x="box",
    marginal_y="violin",
    color="gender",
    trendline="ols",
    opacity=0.9,
    labels={
        "daily_social_media_hours": "Daily Social Media Hours",
        "sleep_hours": "Sleep Hours",
        "gender": "Gender"
    }
)
fig.update_layout(
    title_font_size=20
)
fig.show()
```

```python
#Soru 4: Depresyon riski taşıyan (depression_label = 1) ve taşımayan (0) kişilerin uyku öncesi ekran süreleri (screen_time_before_sleep) arasında belirgin bir fark var mı?
#Önerilen Grafik: Box Plot (Kutu Grafiği) veya Violin Plot - Aykırı değerleri (outliers) görmek için harikadır
fig=px.violin(
    df,
    x="depression_label",
    y="screen_time_before_sleep",
    color="depression_label",
    box=True,
    color_discrete_sequence=["#7d8887", "#380e0e"],
)
fig.update_layout(
    xaxis_title="Effects of depression",
    yaxis_title="Screen Time Before Sleep",
    showlegend=False
)
fig.show()
```

```python
#Soru 5: Cinsiyete (gender) göre stres seviyesi (stress_level) ortalamaları nasıl değişiyor?
#Önerilen Grafik: Bar Plot (ortalama değerler ile)
group_df=df.groupby("gender")["stress_level"].mean().reset_index()
fig=px.bar(
    group_df,
    x="gender",
    y="stress_level",
    color="gender",
    color_discrete_sequence=["#7d8887","#380e0e"],
    title="Average Stress Level by Gender",
)
fig.update_traces(
    marker_line_color="black",
    marker_line_width=2,

)
fig.update_layout(
    xaxis_title="Gender",
    yaxis_title="Average Stress Level",
   
)
fig.show()
```
